### Assignment: Data Science Batch Management System

```python
class UltimateDataScienceBatch:
    def __init__(self,batch_id):
        self.batch_id = batch_id
        self.students = []
    
    def add_student(self,name,email):
        details = {"name": name , "email" : email}
        self.students.append(details)
        print("Added",name,"to",self.batch_id)
    
    def print_student_names(self):
        print("\nStudent List for",self.batch_id + ":")
        for student in self.students:
            print("-",student["name"])

b1 = UltimateDataScienceBatch("Batch 1.0")
b1.add_student("Ankit", "ankit@gmail.com")
b1.add_student("Dinesh","dinesh@gmail.com")

b2 = UltimateDataScienceBatch("Batch 2.0")
b2.add_student("Rohit","rohit@gmail.com")
b2.add_student("Ravi","ravi@gmail.com")
b2.add_student("Atul","atul@gmail.com")

b1.print_student_names()
b2.print_student_names()
```
